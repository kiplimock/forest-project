### Keini Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('kieni_annotation_by_dsail.xlsx', sheet_name='combined')
df.head()
```

```python
species = df["Species"].dropna().unique()
species
```

## Parameter Distribution

### 0. Overall Distribution

```python
df.describe()
```

```python
sns.set(rc={"figure.figsize":(14, 5)})
sns.displot(df, x="CD", kind="kde", fill=True)
plt.title("Crown Diameter")
sns.displot(df, x="TH", kind="kde", fill=True)
plt.title("Tree Height")
plt.show()
```

### 1. Cedar - 'Juniperus procesear'

```python
df_cedar = df[df["Species"] == species[0]]
cd_cedar = df_cedar["CD"]
th_cedar = df_cedar["TH"]

# the df_cedar dataframe has no null values
df_cedar.describe()
```

```python
plt.close()
```

```python
fig, axes = plt.subplots(1, 2)
sns.set(rc={"figure.figsize":(14, 5)})
sns.histplot(df_cedar, x="CD", kde=True, fill=True, ax=axes[0])
axes[0].set_title("Crown Diameter")
sns.histplot(df_cedar, x="TH", kde=True, fill=True, ax=axes[1])
axes[1].set_title("Tree Height")
plt.show()
```
